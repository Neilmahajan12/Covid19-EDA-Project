```python
vaccine_df = pd.read_csv("/content/drive/MyDrive/covid_vaccine_statewise.csv")
vaccine_df.head(10)
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace= True, axis=1)
```

```python
covid_df.head(3)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

```python
covid_df
```

```python
#active cases
covid_df['Active_case']= covid_df["Confirmed"]-(covid_df["Cured"] + covid_df["Deaths"])
covid_df.tail(100)
```

```python
statwise = pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc= max)
```

```python
statwise["RecoveryRate"]= statwise['Cured']*100/statwise['Confirmed']
```

```python
statwise["mortalityRate"]= statwise['Deaths']*100/statwise['Confirmed']
```

```python
statwise = statwise.sort_values(by= 'Confirmed',ascending= False)
```

```python
statwise.style.background_gradient(cmap= 'cubehelix')
```

```python
# Top 10 active cases States
Top_10_active_cases_States=covid_df.groupby(by='State/UnionTerritory').max()[['Active_case','Date']].sort_values(by=['Active_case'],ascending=False).reset_index()
```

```python
fig= plt.figure(figsize=(10,6))
plt.title("Top 10 most active Case in India",size=25)
ax= sns.barplot(data = Top_10_active_cases_States.iloc[:10],y='Active_case',x= 'State/UnionTerritory',linewidth= 2,edgecolor= "blue")
plt.xlabel("States")
plt.ylabel("Total active case")
plt.show()
```

```python
#Top 10 states with highest deaths
top_10_deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig= plt.figure(figsize=(10,6))
plt.title("Top 10 state with highest deaths",size=25)
ax= sns.barplot(data = top_10_deaths.iloc[:12],y='Deaths',x= 'State/UnionTerritory',linewidth= 2,edgecolor= "black")
plt.xlabel("States")
plt.ylabel("Total Death case")
plt.show()
```

```python
#Growth Trend
fig= plt.figure(figsize=(10,6))
ax= sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date', y='Active_case', hue='State/UnionTerritory')
ax.set_title("Top 5 state affected in india",loc='center')
```
